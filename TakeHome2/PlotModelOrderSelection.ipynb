```python
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_theme(style="white")
mpl.use('agg')
mpl.rcParams['font.family'] = 'Linux Libertine'
mpl.rcParams['text.usetex'] = True
mpl.rcParams["font.size"] = 16
sns.set_style("white", {"font.family": "Linux Libertine", "font.size": 18, "figure.figsize": (1.5, 2)})
import matplotlib.pyplot as plt
```

```python
order_selection_kfold = pd.read_csv("/tmp/networkfs-scratch/mat_to_csv_conv_kfold.csv")
order_selection_bic = pd.read_csv("/tmp/networkfs-scratch/mat_to_csv_conv_bic2.csv")
# Add the right column so we know which is which when we concat.
order_selection_bic["SelectionMethod"] = "BIC"
order_selection_kfold["SelectionMethod"] = "KFoldCv"
order_selection = pd.concat([order_selection_kfold, order_selection_bic], axis=0)
```

```python
g = sns.displot(order_selection,
            col="N",
            x="ModelOrder",
            multiple="dodge",
            hue="SelectionMethod",
            stat="probability",
            element="step",
            col_wrap=3,
           )
g.fig.subplots_adjust(top=0.9)
plt.suptitle("Model Order Selection")
plt.savefig("ModelOrder.pdf")
```
